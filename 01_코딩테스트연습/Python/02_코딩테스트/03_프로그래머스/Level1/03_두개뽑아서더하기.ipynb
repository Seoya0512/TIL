```python
from itertools import combinations
import random 

def solution(numbers):
    answer = []

    for i in combinations(numbers, 2):
        add_number = i[0] + i[1]
        answer.append(add_number)
        answer_set = set(answer)
        answer = sorted(list(answer_set))

    return answer
```

```python
numbers = [2,1,3,4,1]
solution(numbers)
```

```python
numbers = [5,0,2,7]
solution(numbers)
```
