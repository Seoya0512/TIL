```python
def solution(s):
    sent = s.split(' ')

    for i in range(len(sent)):         
        word_list = list(sent[i])

        for j in range(len(word_list)):
            if j % 2 == 0:
                word_list[j] = word_list[j].upper()
            else:
                word_list[j] = word_list[j].lower()
        sent[i] = ''.join(word_list)        

    answer = ' '.join(sent) 

    return answer        
```
