### Quiz 1
- makedata 모듈을 이용해서 데이터 프레임 만들기
- user_df
    - 8명의 데이터가 있는 데이터 프레임을 만드세요.
    - UserID
    - Name : makedata.get_name()
    - Age : makedata.get_age()
    - 중복되는 Name 값이 없도록 


```python
names= []
while True:
    name = makedata.get_name()
    if name not in names:
        names.append(name)
    if len(names)>=8 :
        break

datas= {
    'UserID': range(1,9),
    'Name': names ,
    'Age': [makedata.get_age() for _ in range(8)]}

user_df = pd.DataFrame(datas)
```

### Quiz 2
- money_df 만들기
- 15개의 데이터
- ID : 1-8 랜덤한 숫자 데이터
- Money : 1000원 ~ 20000원 데이터까지 숫자로 저장 

```python
import random
# 딕셔너리 데이터를 데이터 프레임에 하나씩 추가하기
money_df = pd.DataFrame(columns=['ID','Money'])

for _ in range(15):
    money_df.loc[len(money_df)]= {
    'ID': random.randint(1,9),
    'Money' :random.randint(1, 21)*1000 
    }
    
# 컬럼데이터에서 Unique 값 확인
ids= money_df['ID'].unique()
ids.sort()
ids
```

```python
money_df
```

```python
# UserID 와 ID 가 같은 값을 가진 경우만 추출 후 합쳐! 
user_df.merge(money_df, left_on='UserID', right_on='ID')
```

```python
# 컬럼명 변경 
user_df.rename(columns={'UserID':'ID'}, inplace=True)
user_df.tail(1)
```

```python
result_df = pd.merge(money_df,user_df)
result_df.tail()
```

```python
# groupby : sum, size, min .. 함수 : Series
money_list = result_df.groupby("Name")['Money'].sum().reset_index()
money_list
```

```python
# groupby : agg("sum"), agg("mean") .. : DataFrame
money_list = result_df.groupby("Name")["Money"].agg("sum").reset_index()
money_list
```

```python
# merge : money_list, user_df = > outer 
# 없는 데이터도 none 으로 나온다 

result = pd.merge(user_df, money_list, how='outer')
result
```

```python
# fillna : NaN을 특정 데이터로 채워주기 

result.fillna(value=0, inplace=True)
result
```

```python
# 데이터별 데이터 타입 확인 
result.dtypes
```
