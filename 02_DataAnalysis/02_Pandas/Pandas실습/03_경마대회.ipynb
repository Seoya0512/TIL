```python
import pandas as pd
import numpy as np
```

##### Race load dataset

```python
# load = encoding
df_race = pd.read_csv("datas/2014_p.csv", encoding="euc-kr")
df_race.tail(2)
```

##### Horse load dataset

```python
df_horse = pd.read_csv('datas/2014_s.csv', encoding='euc-kr')
df_horse.tail(2)
```

### Quiz 1. 어떤 성별을 가진 말이 성적이 좋을까?

- race 데이터 프레임 : 성적데이터
- horse 데이터 프레임 : 성별 데이터
- race 데이터와 horse 데이터를 ID를 key값으로 merge

- 분석 절차
    - 1. 성별에 따른 순위 데이터 프레임으로 생성
    - 2. 전체 데이터에서 성별 데이터를 나눠서 확률 데이터로 변경
    - 3. 1 ~ 5등까지의 성별에 따른 등수 확률 데이터를 출력

```python
df_merged = pd.merge(df_race, df_horse)
df_merged['Count'] = 1
df_merged.head(3)
```

```python
# 성별에 따른 순위 데이터 프레임 생성 

df = df_merged.pivot_table('Count', ['RANK'],['SEX'],aggfunc=np.sum)
df.head(3)
```

```python
# 성별에 대한 전체 데이터
total = df_merged.groupby('SEX')['Count'].sum()
total
```

```python
# 2. 전체 데이터에서 성별 데이터를 나눠서 확률 데이터로 변경
# 소수 둘째자리까지만 출력
# 결론 : 수컷이 좋은 성적을 낼 확률이 높다.

df_result = df / total *100
df_result = round(df_result, 2)
df_result.head()
```

###  Quiz 2. 순위를 승점제로 했을때 2014년 시즌에서 말들의 성적은?

- 1등 : 3점
- 2등 : 2점
- 3등 : 1점

```python
df_merged.tail(2)
```

```python
df_merged['POINT'] = df_merged['RANK'].apply(
    lambda rank : 4 - rank if rank <= 3 else 0)

df_merged.tail(2)
```

```python
df_rank= df_merged.groupby('RCHOSE_NM')['POINT'].agg(sum).reset_index()
df_rank.sort_values('POINT', ascending=False, inplace=True)
df_rank.head(2)
```

```python
df_rank = df_rank.reset_index(drop=True)
df_rank.head()
```
