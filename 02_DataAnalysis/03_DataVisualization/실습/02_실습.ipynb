### 날짜 데이터 추출하기

````python
# 발렌타인데이 기준으로 그래프

import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14' :
            result.append(float(row[-1]))

plt.figure(figsize=(10,2))
plt.title('발렌타인데이의 최고기온 ')
plt.plot(result, 'hotpink')
plt.show()

````

````python
# 1983년 이후 데이터, 최고, 최저 기온 

import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
high=[]
low=[]
average=[]

for row in data :
    if row[-1] != '' and row[-2] != ' ':
        if 1983 <=int(row[0].split('-')[0]):
          if row[0].split('-')[1] =='02' and row[0].split('-')[2] =='14':
            high.append(float(row[-1]))
            low.append(float(row[-2]))
            average.append(float(row[2]))

plt.figure(figsize=(10,2))
plt.title('1983년 이후 최고, 최저기온 ')
plt.plot(high, 'hotpink', label = '최고기온')
plt.plot(low, 'skyblue', label = '최저기온')
plt.plot(average, 'red', label = '평균온도')
plt.legend()
plt.show()
````

### 기온데이터를 히스토그램으로 표현하기

````python
# 최고의 기온 히스토그램 

import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
result= []

for row in data:
  if row[-1] != '':
    result.append(float(row[-1]))

plt.hist(result, bins=100, color='r')
plt.show()
````

````python
# 1월, 8월 최고의 기온 히스토그램 

import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
aug=[]
jan=[]

for row in data:
  month= row[0].split('-')[1]
  if row[-1] != '':
    if month == '08':
      aug.append(float(row[-1]))
    if month == '01':
      jan.append(float(row[-1]))

plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')
plt.legend()
plt.show()
````

#### Box Plot 으로 표현해보기
* 최댓값, 최솟값, 상위 1/4, 2/4(중앙), 3/4에 위치한 값을 보여주는 그래프, 분포를 파악할 수 있음 
* Box = 실수값 분포에서 Q1 과 Q3를 뜻함 
* Box 내부의 가로선 = 중앙값
* IQR(InterQuartile Range)= Q3 - Q1
* outlier 바깥의 점 

````python
#  random 숫자로 박스플롯 그리기

import random
result=[]

for i in range(13):
  result.append(random.randint(1,1000))
print(sorted(result))

plt.boxplot(result)
plt.show()
````

````python
# 1월과 8월의 기온

import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
aug=[]
jan=[]

for row in data:
  month= row[0].split('-')[1]
  if row[-1] != '':
    if month == '08':
      aug.append(float(row[-1]))
    if month == '01':
      jan.append(float(row[-1]))

plt.boxplot([aug,jan])
plt.show()
````

````python
import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
month= [],[],[],[],[],[],[],[],[],[],[],[]

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1]))   # index값 1~12월 -1, 인덱스의 시작은 0

plt.boxplot(month)
plt.show()
````

````python
import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
day=[]
for i in range(31):
  day.append([])

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08':
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.title('8월 기온')
plt.boxplot(day, showfliers=False)
plt.show()
````
