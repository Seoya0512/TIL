##### 서울에 위치한 대학교 5개

```python
unive =['연세대학교', '서울대학교', '고려대학교', '중앙대학교']
```

```python
addr_list= [ ]
for uni in unive:
  option = f'confmKey={api_key}&currentPage=1&countPerPage=10&keyword={quote(uni)}'
  url = f'{road_url}?{option}&resultType=json'
  result = requests.get(url).json()
  addr = result['results']['juso'][0]['roadAddr']
  addr_list.append(addr)

addr_list
```

```python
import pandas as pd

df= pd.DataFrame({
    '대학교' : unive,
    '도로명주소': addr_list
})

df
```

```python
df.to_csv('대학교.csv', index=False)
```

### 카카오 로컬 API 도로명 주소 ==> 위도, 경도

```python
uploaded= files.upload()
filename= list(uploaded.keys())[0]
```

```python
with open(filename) as f:
  api_key = f.read()
len(api_key)
```

```python
addr = '서울특별시 서초구 헌릉로 13(염곡동)'
search_url = "https://dapi.kakao.com/v2/local/search/address.json" 
url = f'{search_url}?query={quote(addr)}'
```

```python
# "Authorization: KakaoAK {REST_API_KEY}"
result= requests.get(url,
                headers={"Authorization": f'KakaoAK {api_key}'}).json()

result       
```

#### 위도, 경도 추출

```python
lng= float(result['documents'][0]['x'])
lat= float(result['documents'][0]['y'])

lng, lat
```

#### 대학교.csv 파일에 위도, 경도 추가하기

```python
uploaded= files.upload()
filename= list(uploaded.keys())[0]
```

```python
df= pd.read_csv(filename)
df
```

#### 반복문 사용하여 각 정보 출력하기

```python
lng_list, lat_list = [],[]

for i in df['도로명주소']:
  url = f'{search_url}?query={i}'
  result= requests.get(url,
                headers={"Authorization": f'KakaoAK {api_key}'}).json()
  lng = float(result['documents'][0]['x'])
  lat = float(result['documents'][0]['y'])
  lng_list.append(lng)
  lat_list.append(lat)
```

```python
df['경도']= lng_list
df['위도']= lat_list
df
```

### 지도 위에 대학교 표기하기

```python
import folium
```

```python
map =folium.Map(location=[df.위도.mean(),df.경도.mean()],zoom_start=11)
for i in df.index[:100]:
  folium.Marker(
      location= [df.위도[i],df.경도[i]],    # 위도, 경도
      popup= folium.Popup(df.도로명주소[i]),
      tooltip= df.대학교[i], 
      icon=folium.Icon(color='green', icon='info-sign')
  ).add_to(map)

title= '<h3 align="center" style="font-size:20px">서울시 대학교안내</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```

### 단계 구분도 (Choropleth)

```python
uploaded= files.upload()
filename= list(uploaded.keys())[0]
```

```python
import pandas as pd

df= pd.read_csv(filename)
df.head()
```

```python
uploaded= files.upload()
filename= list(uploaded.keys())[0]
```

```python
import json
with open(filename) as json_file:
  geo_data= json.load(json_file)
```

```python
geo_data['features'][0]
```

```python
import folium
```

```python
center = [37.581, 126.986]      # 서울 중시부 위도, 경도

map = folium.Map(location=center, zoom_start=11)
folium.TileLayer('cartodbpositron').add_to

folium.Choropleth(
    geo_data= geo_data,         # 지도데이터 파일 경로
    data= df,                   # 시각화 하고자 하는 데이터프레임
    columns=('동','남자'),      # (지도 데이터와 매핑할 값, 시각화 하고자 하는 변수)
    key_on='feature.properties.동',   # 데이터 파일과 매핑할 값 
    fill_color='YlOrRd',                # Color Map
    legend_name= '노령 인구수'        # Color 범주 이름        
).add_to(map)

title= '<h3 align="center" style="font-size:20px">서울시 동별 노령 남성인구</h3>'
map.get_root().html.add_child(folium.Element(title))

map
```
