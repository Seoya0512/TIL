```python
import pandas as pd
import numpy as np

np.random.seed(88)
df = pd.DataFrame({'A': np.linspace(1,10,10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10,4), columns=list('BCDE'))],
               axis =1)
df.iloc[3,3] = np.nan
df.iloc[0,2] = np.nan
df
```

```python
# max, min에 색칠로 표기하기
df.style.highlight_max(color='green').highlight_min(color='pink')
```

```python
# axis =1 (rows)기준으로 색칠 표기하기
df.style.highlight_max(axis=1, color='green').highlight_min(axis=1, color='pink')
```

```python
# null 값 표기하기 (default = red)
df.style.highlight_null()
```

```python
df.style.highlight_null(null_color='green')
```

```python
# null값을 다른 이름으로 변경하노 후, 색칠 표기
df.style.set_na_rep('OutofScope').highlight_null(null_color='orange')
```

```python
# Heatmap 으로 표기
df.style.background_gradient()
```

```python
# css 활용해서 Table Properties 
df.style.set_properties(**{'border':'1.3px balck',
                            'text-align':'center',
                            'color':'green'})
```

```python
# bar chart 그리기
df.style.bar(color =' yellow')
```

```python
# 소수점 2번재 자리까지 출력 
df.style.set_precision(2)
```

```python
df.style.set_caption('This is Datafram styling demo').set_precision(2).background_gradient()
```

```python
# index를 숨기고, 'b'열 을 숨겨라 

df.style.hide_index().hide_columns('B')
```

```python
# format함수를 사용해서 표현 할 수 있다. 소수점 몇번째 자리, % 붙이는 것등등 
# nan에도 붙여나온다는 단점이 있다 

df.style.format("{:.3%}")
```
