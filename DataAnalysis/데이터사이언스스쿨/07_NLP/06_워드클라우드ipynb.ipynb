## 한글 텍스트 전처리

```python
with open(textfile) as fp:
  text = fp.read()
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
tokens = okt.nouns(text)
tokens[:10]
```

```python
# 영문자, 숫자 제거
import re
new_tokens = [ ]
for token in tokens:
  item = re.sub('[A-Za-z0-9]', '', token)
  if item:
    new_tokens.append(item)
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

```python
import nltk
nltk.download('punk')
nltk.download('stopwords')
```

```python
stoptext = """
  선물 추천 것 가격 수 기능 제품 저 제 생각 여자 여자친구 사용 요 더 꽃 구매 고급 
  판매 때 참고 머리 하나 해 한번 제작 용 준 디자인 거 네이버 사람 배송 중 후기 감동
  하트 여친 가능 사랑 분 정말 티니 세대 인기 상품 직접 선택 곳 모양 마감 등 요즘 오늘
  가지 그냥 위 페이 마음 부담 남자 협찬 남녀 전 의미 변경 도움 아래 처 조금 하루
  정도 처리
"""

stop_words = stoptext.split()
new_tokens = [word for word in new_tokens if  word not in stop_words]
new_tokens[:10]
```

```python
gift = nltk.Text(new_tokens, name='여친선물')
plt.figure(figsize=(18,10))
gift.plot(50)
plt.show()
```

### 워드 클라우드 그리기



```python
from wordcloud import WordCloud
from PIL import Image
```

```python
data =gift.vocab().most_common(300)
path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
wc = WordCloud(
    font_path = path, relative_scaling = 0.2,
    background_color = 'white'
).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

* 하트 그림에 표시

```python
mask = np.array(Image.open(maskfile))

from wordcloud import ImageColorGenerator
image_colors = ImageColorGenerator(mask)
```

```python
wc = WordCloud(
    font_path = path, relative_scaling = 0.2,
    background_color = 'white', mask=mask,
    min_font_size =1, max_font_size= 120
).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wc.recolor(color_func=image_colors),
           interpolation='bilinear')
plt.axis('off')
plt.show()
```
