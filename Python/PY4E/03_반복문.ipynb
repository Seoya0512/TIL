```python
import random

def rsp_advanced(games): 
    # ìµœì¢… ê²°ê³¼ ë„ì¶œì„ counting í•˜ê¸° ìœ„í–” ì´ˆê¸°ê°’ ì„¤ì¹˜
    my_win, my_lose = 0,0
    computer_win, computer_lose = 0,0
    draw = 0

    for i in range(games):
        # ì…ë ¥ê°’ì´ ì˜ëª»ë˜ë©´ ì¬ì…ë ¥ ë°›ê¸° 
        while True:
            my = input('ê°€ìœ„ ë°”ìœ„ ë³´ : ')
            if my not in ['0','1','2','ê°€ìœ„','ë°”ìœ„','ë³´']:
                print('Error: ì˜¬ë°”ë¥¸ ê°’ì„ ì…ë ¥í•˜ì„¸ìš”')
                continue 
            else:
                break

        # my ê°’ì„ ìˆ«ìë¡œ ë°›ì€ ê²½ìš° ê¸€ìë¡œ ë³€í™˜ 
        if my == '0':
             my = 'ê°€ìœ„'
        elif my =='1':
            my = 'ë°”ìœ„'
        else:
            my = 'ë³´'

        # ì»´í“¨í„° ê°’ ì…ë ¥ ë°›ê¸° 
        computer = random.randint(0,2)
        if computer == 0:
            computer = 'ê°€ìœ„'
        elif computer == 1:
            computer = 'ë°”ìœ„'
        else:
            computer = 'ë³´'
        
        # ë¹„ê¸´ ê²½ìš°
        if my == computer :
            draw += 1 
            print(f'ë‚˜ : {my}')
            print(f'ì»´í“¨í„° : {computer}')
            print(f'{i+1}ë²ˆì§¸ íŒ ë¹„ê¹€!')
        
        # ë‹¤ë¥¸ ê²½ìš°ì˜ ìˆ˜ 6ê°€ì§€ 
        elif my =='ê°€ìœ„' and computer == 'ë°”ìœ„':
            computer_win += 1
            my_lose += 1
            print(f'ë‚˜ : {my}')
            print(f'ì»´í“¨í„° : {computer}')
            print(f'{i+1}ë²ˆì§¸ íŒ ì»´í“¨í„° ìŠ¹ë¦¬!')
        elif my =='ê°€ìœ„' and computer =='ë³´':
            my_win += 1
            computer_lose += 1
            print(f'ë‚˜ : {my}')
            print(f'ì»´í“¨í„° : {computer}')
            print(f'{i+1}ë²ˆì§¸ íŒ ë‚˜ì˜ ìŠ¹ë¦¬!')
        elif my == 'ë°”ìœ„' and computer =='ê°€ìœ„':
            my_win += 1
            computer_lose += 1
            print(f'ë‚˜ : {my}')
            print(f'ì»´í“¨í„° : {computer}')
            print(f'{i+1}ë²ˆì§¸ íŒ ë‚˜ì˜ ìŠ¹ë¦¬!')
        elif my == 'ë°”ìœ„' and computer == 'ë³´':
            my_lose += 1
            print(f'ë‚˜ : {my}')
            print(f'ì»´í“¨í„° : {computer}')
            print(f'{i+1}ë²ˆì§¸ íŒ ì»´í“¨í„° ìŠ¹ë¦¬!')                       
        elif my == 'ë³´' and computer =='ê°€ìœ„':
            my_lose += 1
            print(f'ë‚˜ : {my}')
            print(f'ì»´í“¨í„° : {computer}')
            print(f'{i+1}ë²ˆì§¸ íŒ ì»´í“¨í„° ìŠ¹ë¦¬!') 
        elif my =='ë³´' and computer == 'ë°”ìœ„':
            my_win += 1
            computer_lose += 1
            print(f'ë‚˜ : {my}')
            print(f'ì»´í“¨í„° : {computer}')
            print(f'{i+1}ë²ˆì§¸ íŒ ë‚˜ì˜ ìŠ¹ë¦¬!')

    print(f'ë‚˜ì˜ ì „ì :{my_win}ìŠ¹ {draw}ë¬´ {my_lose}íŒ¨')
    print(f'ì»´í“¨í„°ì˜ ì „ì : {computer_win}ìŠ¹ {draw}ë¬´ {computer_lose}íŒ¨')           
```

ğŸ“ŒQ4. 2ê°œì˜ ìˆ«ìë¥¼ ì…ë ¥í•˜ì—¬ ê·¸ ì‚¬ì´ì— ì†Œìˆ˜ê°€ ëª‡ ê°œì¸ì§€ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“¤ì–´ ë´…ì‹œë‹¤.

```python
def count_prime_number(n,m):
    count = 0
    for i in range(n,m):
        not_prime = 0 
        if i < 2:
            continue 
        for j in range(2,i):
            if i % j ==0:
                not_prime += 1
        # ë‹¤ë¥¸ìˆ˜ì™€ ë‚˜ëˆ ì§€ëŠ” ê²½ìš°ê°€ ì—†ìœ¼ë©´ ì†Œìˆ˜ 
        if not_prime ==0:
            count += 1
            # ì†Œìˆ˜ ì¶œë ¥ 
            print(i)

    print(f'ì†Œìˆ˜ê°œìˆ˜: {count}')
```

```python
n = int(input("ì²« ë²ˆì§¸ ìˆ˜ ì…ë ¥ : "))
m = int(input("ë‘ ë²ˆì§¸ ìˆ˜ ì…ë ¥ : "))

count_prime_number(n, m)
```
