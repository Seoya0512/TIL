```python
# ì™•ì´ë¦„
korea_king = "íƒœì¡°,í˜œì¢…,ì •ì¢…,ê´‘ì¢…,ê²½ì¢…,ì„±ì¢…,ëª©ì¢…,í˜„ì¢…,ë•ì¢…,ì •ì¢…,ë¬¸ì¢…,ìˆœì¢…,ì„ ì¢…,í—Œì¢…,ìˆ™ì¢…,ì˜ˆì¢…,ì¸ì¢…,ì˜ì¢…,ëª…ì¢…,ì‹ ì¢…,í¬ì¢…,ê°•ì¢…,ê³ ì¢…,ì›ì¡°,ì¶©ë ¬ì™•,ì¶©ì„ ì™•,ì¶©ìˆ™ì™•,ì¶©í˜œì™•,ì¶©ëª©ì™•,ì¶©ì •ì™•,ê³µë¯¼ì™•,ìš°ì™•,ì°½ì™•,ê³µì–‘ì™•"
chosun_king = "íƒœì¡°,ì •ì¢…,íƒœì¢…,ì„¸ì¢…,ë¬¸ì¢…,ë‹¨ì¢…,ì„¸ì¡°,ì˜ˆì¢…,ì„±ì¢…,ì—°ì‚°êµ°,ì¤‘ì¢…,ì¸ì¢…,ëª…ì¢…,ì„ ì¡°,ê´‘í•´êµ°,ì¸ì¡°,íš¨ì¢…,í˜„ì¢…,ìˆ™ì¢…,ê²½ì¢…,ì˜ì¡°,ì •ì¡°,ìˆœì¡°,í—Œì¢…,ì² ì¢…,ê³ ì¢…,ìˆœì¢…"
```

```python
def king(korea_king,chosun_king):
    name_list= []
    korea = korea_king.split(',')           # ê³ ë ¤ì™• ë¦¬ìŠ¤íŠ¸ ë§Œë“¤ê¸°
    chosun = chosun_king.split(',')         # ì¡°ì„ ì‹œëŒ€ ì™• ë¦¬ìŠ¤íŠ¸ ë§Œë“¤ê¸°
    
    # ì¤‘ë³µì´ ë˜ì§€ ì•ŠëŠ” ì™• ë¦¬ìŠ¤íŠ¸ ë§Œë“¤ê¸° 
    for kor in korea:
        for cho in chosun:
            if kor == cho :
                name = kor
                name_list.append(name)

    # ë¦¬ìŠ¤íŠ¸ì—ì„œ ì¤‘ë³µì„ ì œê±°í•˜ëŠ” dict.fromkeys ì‚¬ìš©í•´ì„œ í•œë²ˆ ë” ì •ë¦¬          
    kings= dict.fromkeys(name_list)
    king_list= list(kings)
    king_list

    count = 0
    for king in king_list:
        print(f'ì¡°ì„ ê³¼ ê³ ë ¤ì— ëª¨ë‘ ìˆëŠ” ì™• ì´ë¦„: {king}')
        count += 1
    print(f'ì¡°ì„ ê³¼ ê³ ë ¤ì— ëª¨ë‘ ìˆëŠ” ì™• ì´ë¦„ì€ ì´ {count}ê°œ ì…ë‹ˆë‹¤')
```

ğŸ˜²ì¡°ê±´ 4 - ë©´ë‹´ ëŒ€ìƒìì˜ í‰ê·  ì‹¤ì ì´ 3ë³´ë‹¤ í¬ë©´ ë©´ë‹´ ëŒ€ìƒìì—ì„œ ì œì™¸ ë©ë‹ˆë‹¤.

```python
# ì´ë¦„, ì‹¤ì 
member_names = ["ê°‘ëŒì´", "ê°‘ìˆœì´", "ì„ëŒì´", "ì„ìˆœì´", "ë³‘ëŒì´", "ë³‘ìˆœì´"]
member_records = [[4,5,3,5,6,5,3,4,1,3,4,5],[2,3,4,3,1,2,0,3,2,5,7,2], 
           [1,3,0,3,3,4,5,6,7,2,2,1],[3,2,9,2,3,5,6,6,4,6,9,9],
           [8,7,7,5,6,7,5,8,8,6,10,9],[7,8,4,9,5,10,3,3,2,2,1,3]]
```

```python
# ì§ì›ë³„ í‰ê·  ì‹¤ì  êµ¬í•˜ê¸° 
mean_list= []

def sales_management(member_names, member_records):
    for i in range(len(member_records)):                                # í–‰ì„ ë¶ˆëŸ¬ì˜¤ê¸° 
        mean= sum(member_records[i])/len(member_records[i])
        mean_list.append(mean)                                          # í‰ê·  ë¦¬ìŠ¤íŠ¸ êµ¬í•˜ê¸° 
     
     # recrd_dict() ë§Œë“¤ì–´ ì •ë¦¬ {name: mean}
    record_dict= dict(zip(member_names, mean_list))
     # value ê¸°ì¤€ìœ¼ë¡œ ranking ì •ë¦¬             
    ranking = [ (v,k) for k,v in record_dict.items() ]           
    ranking = sorted(ranking, reverse=True)

    # ë³´ë„ˆìŠ¤ ëŒ€ìƒì êµ¬í•˜ê¸° 
    bonus_names = (ranking[0][1], ranking[1][1])
    for i in bonus_names:
        if record_dict[i] < 5:
            continue
        print(f"ë³´ë„ˆìŠ¤ ëŒ€ìƒì {i}")
    print()

    # ë©´ë‹´ ëŒ€ìƒì êµ¬í•˜ê¸°
    counsel_name = (ranking[4][1], ranking[5][1])
    for k in counsel_name:
        if record_dict[k] > 3:
            continue
        print(f'ë©´ë‹´ ëŒ€ìƒì {k}')
    print()
```
