```python
import re 

def solution(s):
    p_1= re.compile('[0-9+-]')
    p_2 = re.compile('[0-9]')

    answer = ''
    for i in range(len(s)):
        if i == 0:
            m = p_1.match(s[i]).group()
        else: 
            m = p_2.match(s[i]).group()
        answer += m

    return int(answer)
```

```python
s = '-12345'
solution(s)
```
